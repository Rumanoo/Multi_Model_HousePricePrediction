```python
import os
import pandas as pd
import numpy as np
from PIL import Image

import torch
import torch.nn as nn
from torchvision import models, transforms

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
CSV_PATH = r"D:\Datasets\House\socal2.csv"
IMAGE_DIR = r"D:\Datasets\House\socal_pics"
```

```python
df = pd.read_csv(CSV_PATH)
df.head()
```

```python
tabular = df[['bed', 'bath', 'sqft']]
target = df['price']
image_ids = df['image_id'].astype(str)

scaler = StandardScaler()
tabular_scaled = scaler.fit_transform(tabular)
```

```python
img_transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
```

```python
image_tensors = []
valid_idx = []

for i, img_id in enumerate(image_ids):
    img_path = os.path.join(IMAGE_DIR, f"{img_id}.jpg")
    if os.path.exists(img_path):
        img = Image.open(img_path).convert("RGB")
        image_tensors.append(img_transform(img))
        valid_idx.append(i)

images_tensor = torch.stack(image_tensors)

tabular_scaled = tabular_scaled[valid_idx]
target = target.iloc[valid_idx].reset_index(drop=True)
```
